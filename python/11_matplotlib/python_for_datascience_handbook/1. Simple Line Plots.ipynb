### Simple Line Plots

• Perhaps the simplest of all plots is the visualization of a single function y = f(x).

• Here we will take a first look at creating a simple plot of this type.

• As with all the following sections, we’ll start by setting up the notebook for plotting and importing the functions we will use.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

• For all Matplotlib plots, we start by creating a figure and an axes.

• In their simplest form, a figure and axes can be created as follows.

```python
fig = plt.figure()
axis = plt.axes()
plt.title("An empty gridded axes")
```

• Once we have created an axes, we can use the ax.plot function to plot some data.

• Let’s start with a simple sinusoid.

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
plt.title('A simple sinusoid');
```

• Alternatively, we can use the pylab interface and let the figure and axes be created for us in the background.

```python
plt.title("A simple sinusoid via the object-oriented interface")
plt.plot(x, np.sin(x));
```

• If we want to create a single figure with multiple lines, we can simply call the plot function multiple times.

```python
plt.title('Over-plotting multiple lines')
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

### Adjusting the Plot: Line Colors and Styles

• The first adjustment you might wish to make to a plot is to control the line colors and styles.

• The plt.plot() function takes additional arguments that can be used to specify these.

• To adjust the color, you can use the color keyword, which accepts a string argument representing virtually any imaginable color.

• The color can be specified in a variety of ways.

```python
plt.title('Controlling the color of plot elements')
plt.plot(x, np.sin(x - 0), color = 'black') # Specify color by name
plt.plot(x, np.sin(x - 1), color = 'g') # Short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color = '0.75') # Gray scale between 0 and 1
plt.plot(x, np.sin(x - 3), color = '#FFDD44') # Hex Code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color = (0.5, 0, 0)) # RGB tuple, values 0 and 1
plt.plot(x,np.sin(x - 5), color='chartreuse'); # all HRML color names supported
```

• If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.

```python
plt.title('Controlling the color of plot elements')
plt.plot(x, np.sin(x - 0)) # Specify color by name
plt.plot(x, np.sin(x - 1)) # Short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2)) # Gray scale between 0 and 1
plt.plot(x, np.sin(x - 3)) # Hex Code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4)) # RGB tuple, values 0 and 1
plt.plot(x,np.sin(x - 5)); # all HRML color names supported
```

• Similarly, you can adjust the line style using the linestyle keyword.

```python
plt.title('Example of various line styles')
plt.plot(x, x + 1, linestyle = 'solid')
plt.plot(x, x + 2, linestyle = 'dashed')
plt.plot(x, x + 3, linestyle = 'dotted')
plt.plot(x, x + 4, linestyle = 'dashdot')

# For short, we can use the following codes
plt.plot(x, x + 5, linestyle = '-')
plt.plot(x, x + 6, linestyle = '--')
plt.plot(x, x + 7, linestyle = ':')
plt.plot(x, x + 8, linestyle = '-.');
```

• If you would like to be extremely terse, these linestyle and color codes can be combined into a single nonkeyword argument to the plt.plot() function

```python
plt.title('Controlling colors and styles with the shorthand syntax')
plt.plot(x, x + 0, '-g') # Solid Green
plt.plot(x, x + 1, '--c') # Dashed Cyan
plt.plot(x, x + 2, '-.k') # Dashdot Black
plt.plot(x, x + 3, ':r'); # Dotted Red
```

• These single-character color codes reflect the standard abbreviations in the RG(Red/Green/Blue) and CMYK (Cyan/Magenta/Yellow/blacK) color systems, commonly used for digital color graphics.

### Adjusting the Plot: Axes Limits

• Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes it’s nice to have finer control.

• The most basic way to adjust axis limits is to use the plt.xlim() and plt.ylim() methods.

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 12)
plt.ylim(-1.5, 1.5)
plt.title('Example of axis limits');
```

• If for some reason you’d like either axis to be displayed in reverse, you can simply reverse the order of the arguments

```python
plt.plot(x, np.sin(x))
plt.plot(10, 1)
plt.plot(1.2, -1.2)
plt.plot('Example of reversing the y-axis');
```

• A useful related method is plt.axis() (note here the potential confusion between axes with an e, and axis with an i).

• The plt.axis() method allows you to set the x and y limits with a single call, by passing a list that specifies [xmin, xmax, ymin,ymax].

```python
plt.plot(x, np.sin(x))
plt.axis([-5, 10, -1.8, 2.5])
plt.title('Setting the axis limits with plt.axis');
```

• The plt.axis() method goes even beyond this, allowing you to do things like automatically tighten the bounds around the current plot.

```python
plt.plot(x, np.sin(x))
plt.title('Example of a "tight" layout')
plt.axis('tight');
```

• It allows even higher-level specifications, such as ensuring an equal aspect ratio so that on your screen, one unit in x is equal to one unit in y.

```python
plt.plot(x, np.sin(x))
plt.title('Example of an “equal” layout, with units matched to the output resolution')
plt.axis('equal');
```

### Labeling Plots

• As the last piece of this section, we’ll briefly look at the labeling of plots: titles, axis labels, and simple legends.

• Titles and axis labels are the simplest such labels—there are methods that can be used to quickly set them.

```python
plt.plot(x, np.sin(x))
plt.title('A sin Curve')
plt.xlabel('x')
plt.ylabel('sin(x)');
```

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')
plt.title('Plot legend example')
plt.legend();
```
