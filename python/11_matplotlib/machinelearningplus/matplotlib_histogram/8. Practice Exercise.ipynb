```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import data
df = sns.load_dataset('iris')
df.head()
```

```python
plt.subplots(figsize = (7, 6), dpi = 100)

sns.distplot(df.loc[df.species == 'setosa', 'sepal_length'], label = 'setosa', color = 'dodgerblue')
sns.distplot(df.loc[df.species == 'versicolor', 'sepal_length'], label = 'versicolor', color = 'orange')
sns.distplot(df.loc[df.species == 'virginica', 'sepal_length'], label = 'virginica', color = 'deeppink')

plt.title('Iris Histogram')
plt.legend();
```
