### How to Customise the Legend

• The most common way to make a legend is to define the label parameter for each of the plots and finally call plt.legend().

• However, sometimes you might want to construct the legend on your own.

• In that case, you need to pass the plot items you want to draw the legend for and the legend text as parameters to plt.legend() in the following format:

    plt.legend((line1, line2, line3), ('label1', 'label2', 'label3'))

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn-notebook')

plt.figure(figsize = (10, 7), dpi = 180)
X = np.linspace(0, 2 * np.pi, 1000)
sin1 = plt.plot(X, np.sin(X))
cos1 = plt.plot(X, np.cos(X))
sin2 = plt.plot(X, np.sin(X + 0.5))
cos2 = plt.plot(X, np.cos(X + 0.5))
plt.gca().set(ylim = (-1.25, 1.5), xlim = (-0.5, 7))
plt.title('Custom Legend Example', fontsize = 10)

# Modifiy Legend
plt.legend([sin1[0], cos1[0], sin2[0], cos2[0]], # Plot items
          ['sin1 curve', 'cos1 curve', 'sin2 curve', 'cos2 curve'],
          frameon = True, # Plot legend border
          framealpha = 1, # Transperancy of border
          ncol = 2, # number of columns
          shadow = True, # Shadow on
          borderpad = 1, # Thickness of border
          title = 'Sins & Cosines');
```
