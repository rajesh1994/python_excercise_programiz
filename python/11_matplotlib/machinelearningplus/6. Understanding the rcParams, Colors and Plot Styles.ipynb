### Understanding the rcParams, Colors and Plot Styles

• You can adjust the params you’d like to change by updating it.

• The below snippet adjusts the font by setting it to ‘stix’, which looks great on plots by the way.

```python
import matplotlib.rcsetup as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
mpl.rcParams.update({'font.size' : 18, 'font.family' : 'STIXGeneral', 'mathtext.fontset' : 'stix'})
```

• After modifying a plot, you can rollback the rcParams to default setting using:

```python
# reset to defaults
mpl.rcParams.update(mpl.rcParamsDefault)
```

• Matplotlib comes with pre-built styles which you can look by typing:

```python
plt.style.available
```

```python
import matplotlib as mpl

mpl.rcParams.update({'font.size' : 18, 'font.family' : 'STIXGeneral', 'mathtext.fontset' : 'stix'})

def plot_sine_cosine_wave(style = 'ggplot'):
    plt.style.use(style)
    plt.figure(figsize = (7, 4), dpi = 80)
    X = np.linspace(0, 2 * np.pi, 1000)
    plt.plot(X, np.sin(X))
    plt.plot(X, np.cos(X))
    plt.xticks(ticks=np.arange(0, 440/57.2985, 90/57.2985),
               labels = [r'$0$',r'$\frac{\pi}{2}$',r'$\pi$',r'$\frac{3\pi}{2}$',r'$2\pi$']) # 1 radian = 57.2985 degrees
    plt.gca().set(ylim = (-1.25, 1.25), xlim = (-.5, 7))
    plt.title(style, fontsize = 18)
    plt.show();
    
    plot_sine_cosine_wave('seaborn-notebook')
    plot_sine_cosine_wave('ggplot')
    plot_sine_cosine_wave('bmh')
```
