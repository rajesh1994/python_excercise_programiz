### Introduction to Seaborn
• As the charts get more complex, the more the code you’ve got to write.

• For example, in matplotlib, there is no direct method to draw a density plot of a scatterplot with line of best fit.

• You get the idea.

• So, what you can do instead is to use a higher level package like seaborn, and use one of its prebuilt functions to draw the plot.

• We are not going in-depth into seaborn. But let’s see how to get started and where to find what you want.

• A lot of seaborn’s plots are suitable for data analysis and the library works seamlessly with pandas dataframes.

• seaborn is typically imported as sns. Like matplotlib it comes with its own set of pre-built styles and palettes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the data
data = pd.read_csv("mpg_ggplot2.csv")

# Draw plot
plt.figure(figsize = (16, 10), dpi = 80)
sns.kdeplot(data.loc[data['cyl'] == 4, 'cty'], shade = True, color = 'g', label = 'Cyl = 4', alpha = 0.7)
sns.kdeplot(data.loc[data['cyl'] == 6, 'cty'], shade = True, color = 'dodgerblue', label = 'Cyl = 6', alpha = 0.7)
sns.kdeplot(data.loc[data['cyl'] == 8, 'cty'], shade = True, color = 'orange', label = 'Cyl = 8', alpha = 0.7)

# Decoration
plt.title('Density plot of city mileage by n_Cylinders', fontsize = 16)
plt.legend();
```

```python
# Load Dataset
data = sns.load_dataset('iris')

# Plot
plt.figure(figsize = (10, 8), dpi = 8)
sns.pairplot(data, kind = 'reg', hue = 'species');
```
