```python
# Importing the libraries
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date
import numpy as np
```

```python
# Reading CSV File
df = pd.read_csv("/home/rajesh/Documents/Motorq/24-June-2019/trip_feed/trip_feed_05.csv")
df.head()
```

```python
# Replacing the strings
modified_startTimestamp = df["startTimestamp"].str.replace("T", " ").str.replace("Z", "")
modified_endTimestamp = df["endTimestamp"].str.replace("T", " ").str.replace("Z", "")
```

```python
# Changing the datatype
df["modified_startTimestamp"] = pd.to_datetime(modified_startTimestamp)
df["modified_endTimestamp"] = pd.to_datetime(modified_endTimestamp)
df.head()
```

```python
df["time_diff"] = df['modified_endTimestamp'] - df['modified_startTimestamp']

# Time difference in minutes
df["time_diff_in_mins"] = round(df["time_diff"]/np.timedelta64(1,'m'), 2)

# Time difference in hours
df["time_diff_in_hours"] = round(df["time_diff"]/np.timedelta64(1,'h'), 2)

# Time difference in seconds
df["time_diff_in_sec"] = df["time_diff"]/np.timedelta64(1,'s')

# Droping unwanted columns
df.drop(['modified_startTimestamp', 'modified_endTimestamp', 'time_diff'], inplace=True, axis="columns")

# Exporting the dataframe as CSV file
df.to_csv(r"/home/rajesh/Documents/Motorq/24-June-2019/trip_feed/trip_feed_(time_diff_in_mins)_05.csv", index=False)

# Printing the dataframe
df.head()
```
